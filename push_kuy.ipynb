# Caranya:


1. copy ssh **link**
2. **git clone** 'ssh link' -> outputnya akan membuat folder baru
3. cd ke new folder tersebut
4. **git status** -> outputnya akan mencul nama-nama file di folder baru tsb
5. *git add. * -> artinya akan ngepush seluruh file di folder tsb

**bisa juga dit add <nama-file> -> ngepush file tertentu**

6. **git status** -> outputnya akan muncul nama-nama di file folder baru yg sudah di add berwarna hijau
7. **git commit** -m "revisi_1"
8. 'git push'
